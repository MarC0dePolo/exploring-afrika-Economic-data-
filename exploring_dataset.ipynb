```python

for country in df['country'].unique():
    plt.figure(figsize=(12, 6))
    country_data = df[df['country'] == country]
    
    inf_crises = country_data[country_data['inflation_crises'] == 1]
    bank_crises = country_data[country_data['banking_crisis'] == 1]
    sys_crisis = country_data[country_data["systemic_crisis"] == 1]
    
    plt.plot(country_data['year'], country_data['inflation_annual_cpi'], label='Inflation Rate', color='black', linestyle='--', markersize=4, alpha=0.7)
    

    plt.scatter(inf_crises['year'], inf_crises['inflation_annual_cpi'], color='red', edgecolor='red', marker='s', linewidth=3, label='Inflation Crisis')
    plt.scatter(bank_crises['year'], bank_crises['inflation_annual_cpi'], color='blue', edgecolor='blue', marker='v', label='Banking Crisis')
    plt.scatter(sys_crisis['year'], sys_crisis['inflation_annual_cpi'], color='green', edgecolor='green', marker='^', label='Systemic Crisis')
    
    
    plt.axhline(color="grey", linestyle="--", linewidth=1)

    
    plt.xlabel('Year')
    plt.ylabel('Annual CPI Inflation')
    plt.title(f"{country} Inflation rate & Occouring Crises")
    plt.legend()
    plt.grid(True)
    plt.show()

```

```python
for c in df['country'].unique():
    c_data = df[df['country'] == c]
    dom_default = c_data[c_data['domestic_debt_in_default'] == 1]

    if len(dom_default) < 1:
        continue

    plt.figure(figsize=(12,6))
    
    #plot for Inflation Data
    plt.plot(c_data["year"], c_data["inflation_annual_cpi"], linestyle="-", color="black", label="Inflation Rate")
    plt.xlabel("Year")
    plt.ylabel("Inflation Rate")

    plt.axhline(y=0, linestyle='--', color="black", linewidth=1)

    plt.scatter(dom_default['year'], dom_default['inflation_annual_cpi'], marker='o', color="red", label="Domestic Default Occoured")
    
    plt.title(f"{country} USD Exchange Rate & Domestic Default")

    plt.legend()
    plt.grid(True)
    plt.show()

### All countries were a Domestic Dept Default occoured
```

```python
for country in df['country'].unique():
    plt.figure(figsize=(12, 6))
    
    country_data = df[df['country'] == country]
    curr_crisis = country_data[country_data['currency_crises'] == 1]
    
    plt.plot(country_data['year'], country_data['exch_usd'], label='USD Exchange Rate', color='black', linestyle='--', markersize=4)

    plt.scatter(curr_crisis['year'], curr_crisis['exch_usd'], color='yellow', edgecolor='black', marker='s', label='Currency Crisis')

    plt.axhline(color="grey", linestyle="--", linewidth=1)
    
    plt.xlabel('Year')
    plt.ylabel('USD Exchange Rate')
    plt.title(f"{country} USD Exchange Rate & Occouring Currency Crisis")
    plt.legend()
    plt.grid(True)
    plt.show()

```

```python
for c in df['country'].unique():
    c_data = df[df['country'] == c]
    sov_default = c_data[c_data['sovereign_external_debt_default'] == 1]

    if len(sov_default) < 1:
        continue

    fig, ax1 = plt.subplots(figsize=(12,6))
    ax1.plot(c_data["year"], c_data["inflation_annual_cpi"], linestyle="-", color="tab:blue", label="Inflation Rate")
    ax1.set_xlabel("Year")
    ax1.set_ylabel("Inlfation Rate", color="tab:blue")
    ax1.tick_params(axis='y', labelcolor="tab:blue")
    ax1.scatter(sov_default['year'], sov_default['inflation_annual_cpi'], marker='o', color="black")
    
    ax1.axhline(linestyle="--", color="tab:blue", linewidth=1)
    
    
    ax2 = ax1.twinx()
    ax2.plot(c_data["year"], c_data["exch_usd"], linestyle="-", color="tab:green", label="USD Exchange Rate")
    ax2.set_ylabel("USD Exchange Rate", color="tab:green")
    ax2.tick_params(axis='y', labelcolor="tab:green")
    ax2.scatter(sov_default['year'], sov_default['exch_usd'], marker='o', color="black", label=" Occouring Sovereign Dept Default")

    ax2.axhline(linestyle="--", color="tab:green", linewidth=1)

    plt.grid(True)
    plt.tight_layout()
    
    lines, labels = ax1.get_legend_handles_labels()
    lines2, labels2 = ax2.get_legend_handles_labels()

    ax2.legend(lines + lines2, labels + labels2, loc='upper left')
    plt.title(f"{c} Correlation Between USD Exchange Rate Inflation And Occouring Sov Dept Default")
    
    plt.show()

    ### I wanted to see if there is a higher correlation in
    ### an occouring Sov Dept Default and a increasing inflation,
    ### or a occouring Sov Dept Default and a increasing USD Exchange rate

    ### For Angola, Nigeria, Zambia it seems like making more sense to plot Inflation with the occouring Sovereign Dept Default
```
